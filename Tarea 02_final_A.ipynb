# Desarrollo
## 1) Preprocesamiento de Datos


```python
#cargamos las librerias necesarias para nuestros analisis
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.linear_model import LinearRegression
from scipy.stats import f_oneway, chi2_contingency

# Configuración inicial para las figuras
plt.rcParams['figure.figsize'] = (10, 6)
```

```python
#Importación del dataset:
# Cargar los datos
#De acuerdo a las indicaciones y sugerencias usamos encoding="cp1252" 
data = pd.read_csv("base-enusc-2023.csv", encoding="cp1252")

# Verificar datos faltantes por columna
print("Datos faltantes por columna:")
print(data.isnull().sum())

# Eliminar duplicados
data_clean = data.drop_duplicates()

#seelcionar las VARIABLES DE INTERES
variables_interes = [
    'enc_region', 'rph_edad', 'rph_nse', 
    'P_AUMENTO_PAIS', 'P_INSEG_LUGARES_1', 'P_INSEG_LUGARES_2',
    'SCREEN_ROB_RVI', 'SCREEN_ROB_HUR', 'EV_CONFIA_CCH','POSESION_ARMA'
]

dat2 = data_clean[variables_interes]
dat2 = dat2[dat2['rph_nse'].isin([1, 2, 3, 4, 5])]

#reemplzar datos faltantes si en este punto todavia se presentaban 
imputer = SimpleImputer(strategy="mean")
dat2 = pd.DataFrame(imputer.fit_transform(dat2), columns=variables_interes)
```

## 2) Análisis Exploratorio

```python
# Estadísticos descriptivos (esatdisticos basicos por cada columna)
print("Variables seleccionadas:")
print(dat2.describe())

# Graico de la distribución de encuestas por región
plt.figure(figsize=(12, 6))
data_region = dat2['enc_region'].value_counts()
plt.bar(data_region.index, data_region.values, color="skyblue")
plt.title("Distribución de encuestas por región")
plt.xlabel("Región")
plt.ylabel("Frecuencia")
#plt.savefig("encuestas por región.jpg", bbox_inches='tight')
plt.show()

# Graficar la matriz de correlación entre las variables numéricas seleccionadas
numeric_columns = ['P_AUMENTO_PAIS', 'P_INSEG_LUGARES_1', 'P_INSEG_LUGARES_2', 'SCREEN_ROB_RVI', 'SCREEN_ROB_HUR']
correlation_matrix = dat2[numeric_columns].corr()

plt.figure(figsize=(8, 6))
plt.imshow(correlation_matrix, cmap="coolwarm", interpolation='none')
plt.colorbar(label="Correlación")
plt.xticks(range(len(numeric_columns)), numeric_columns, rotation=45)
plt.yticks(range(len(numeric_columns)), numeric_columns)
# Agregar valores de correlación en cada celda
for i in range(len(numeric_columns)):
    for j in range(len(numeric_columns)):
        plt.text(j, i, f"{correlation_matrix.iloc[i, j]:.2f}",
                 ha="center", va="center", color="black")
plt.title("Matriz de Correlación")
#plt.savefig("CorrelaciOn.jpg", bbox_inches='tight')
plt.show()

# Grafico de barras de perpsepcion de seguirdad por Regióm y nivel socioeconómico
plt.figure(figsize=(12, 6))
data_region_nse = dat2.groupby(['enc_region', 'rph_nse'])['POSESION_ARMA'].mean().unstack()
data_region_nse.plot(kind="bar", colormap="viridis")
plt.title("Percepción de inseguridad por región y nivel socioeconómico")
plt.xlabel("Región")
plt.ylabel("Percepción de inseguridad")
plt.legend(title="Nivel Socioeconómico")
#plt.savefig("inseguridad por región y nivel socioeconómico.jpg", bbox_inches='tight')
plt.show()

# Grafico comparación de percepción de inseguridad por edad y nivel socioeconómico
plt.figure(figsize=(12, 6))
data_age_nse = dat2.groupby(['rph_edad', 'rph_nse'])['P_INSEG_LUGARES_1'].mean().unstack()
data_age_nse.plot(kind="line", marker="o", colormap="viridis")
plt.title("Percepción de inseguridad según edad y nivel socioeconómico")
plt.xlabel("Edad")
plt.ylabel("Percepción de inseguridad")
plt.legend(title="Nivel Socioeconómico")
#plt.savefig("inseguridad según edad y nivel socioeconómico.jpg", bbox_inches='tight')
plt.show()

# Percepción de inseguridad por región y nivel socioeconómico
plt.figure(figsize=(12, 6))
data_region_nse = dat2.groupby(['enc_region', 'rph_nse'])['P_INSEG_LUGARES_1'].mean().unstack()
data_region_nse.plot(kind="bar", stacked=True, colormap="viridis")

plt.title("Percepción de inseguridad según la región y nivel socioeconómico")
plt.xlabel("Región")
plt.ylabel("Percepción de inseguridad")
plt.legend(title="Nivel Socioeconómico")
#plt.savefig("inseguridad según la región y nivel socioeconómico.jpg", bbox_inches='tight')
plt.show()
```

## 3) Análisis Estadísticos

```python
#Estadisticos adicionales
#ANOVA-Comparar posesión de armas entre niveles socioeconómicos
anova_nse = f_oneway(*[dat2[dat2['rph_nse'] == n]['POSESION_ARMA'] for n in dat2['rph_nse'].unique()])
print(f"ANOVA (posesión de armas por nivel socioeconómico): F={anova_nse.statistic:.2f}, p={anova_nse.pvalue:.4f}")

#ANOVA-Comparar posesión de armas entre regiones
anova_region = f_oneway(*[dat2[dat2['enc_region'] == r]['POSESION_ARMA'] for r in dat2['enc_region'].unique()])
print(f"ANOVA (posesión de armas por región): F={anova_region.statistic:.2f}, p={anova_region.pvalue:.4f}")

#ANOVA de dos factores: Región y Nivel Socioeconómico
# Crear lista de grupos para el ANOVA de dos vías
groups = []
for n in dat2['rph_nse'].unique():  # Para cada nivel socioeconómico
    for r in dat2['enc_region'].unique():  # Para cada región
        group = dat2[(dat2['rph_nse'] == n) & (dat2['enc_region'] == r)]['POSESION_ARMA']
        groups.append(group)

# Realizar el ANOVA de dos vías usando f_oneway
anova_result = f_oneway(*groups)

# Mostrar resultados
print(f"ANOVA de dos vías: F={anova_result.statistic:.2f}, p={anova_result.pvalue:.4f}")


# Chi-cuadrado: Asociaciones entre región y nivel socioeconómico
contingencia = pd.crosstab(dat2['enc_region'], dat2['rph_nse'])
chi2, p, dof, expected = chi2_contingency(contingencia)
print(f"Chi-cuadrado: χ²={chi2:.2f}, p={p:.4f}")

##Regresión lineal
X = dat2[['rph_edad', 'rph_nse']]
y = dat2['P_INSEG_LUGARES_1']

# Modelo de regresión lineal
reg = LinearRegression()
reg.fit(X, y)

# Coeficientes
print("Coeficientes de la regresión lineal:")
for col, coef in zip(X.columns, reg.coef_):
    print(f"{col}: {coef:.4f}")
print(f"Intercepto: {reg.intercept_:.4f}")

# Visualización de la regresión lineal
plt.figure(figsize=(12, 6))
plt.scatter(X['rph_edad'], y, color='blue', label='Datos reales', alpha=0.5)
plt.plot(X['rph_edad'], reg.predict(X), color='red', label='Línea de regresión')
plt.title("Regresión Lineal: Percepción de inseguridad vs Edad")
plt.xlabel("Edad")
plt.ylabel("Percepción de inseguridad (P_INSEG_LUGARES_1)")
plt.legend()
plt.show()
```

## 4) Modelos Predictivos
## Aplicación de Algoritmos de Machine Learning


```python
### KMeans para identificar grupos

scaler = StandardScaler()
data_scaled = scaler.fit_transform(dat2[numeric_columns])

range_clusters = range(2, 6)
silhouette_scores = []

for k in range_clusters:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = kmeans.fit_predict(data_scaled)
    score = silhouette_score(data_scaled, labels)
    silhouette_scores.append(score)
    print(f"K={k}, Silhouette Score: {score:.4f}")

# Elegir el mejor número de clusters
best_k = range_clusters[np.argmax(silhouette_scores)]
print(f"\nNúmero óptimo de clusters: {best_k}")

# Aplicar KMeans
kmeans = KMeans(n_clusters=best_k, random_state=42, n_init=10)
dat2['Cluster'] = kmeans.fit_predict(data_scaled)

# Visualizar clusters
plt.figure(figsize=(12, 6))
plt.scatter(dat2['P_INSEG_LUGARES_1'], dat2['P_INSEG_LUGARES_2'], c=dat2['Cluster'], cmap="viridis", alpha=0.6)
plt.colorbar(label="Cluster")
plt.title("Clusters identificados con KMeans")
plt.xlabel("Inseguridad en vehículo")
plt.ylabel("Inseguridad en transporte público")
plt.show()

# Análisis descriptivo por cluster
cluster_stats = dat2.groupby('Cluster').mean()
print("Estadísticas descriptivas por cluster:")
print(cluster_stats)

# Visualización de la distribución de una variable por cluster
plt.figure(figsize=(12, 6))
for cluster in dat2['Cluster'].unique():
    subset = dat2[dat2['Cluster'] == cluster]
    plt.hist(subset['P_INSEG_LUGARES_1'], alpha=0.5, label=f"Cluster {cluster}")

plt.title("Distribución de percepción de inseguridad por cluster")
plt.xlabel("Percepción de inseguridad (P_INSEG_LUGARES_1)")
plt.ylabel("Frecuencia")
plt.legend()
plt.show()
```
